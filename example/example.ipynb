```python
import ganttpy as gp
```

```python
# Define the tasks we want to plot
tasks = [
    gp.Task('Task A', start='Jan 10, 2025', finish='Feb 20, 2025', color='tab:blue'),
    gp.Task('Task B', start='Feb 1, 2025', finish='Apr 1, 2025', color='tab:blue'),
    gp.Task('Task C', start='Feb 20 2025', duration=21, color='tab:red'),
    gp.Task('Task D', start='Apr 1, 2025', finish='Apr 30, 2025', color='tab:green'),
    gp.Task('Task E', start='Apr 15, 2025', finish='May 15, 2025', color='tab:orange'),
    gp.Task('Task F', start='May 8, 2025', finish='Jun 1, 2025', color='tab:orange'),
    gp.Task('Task G', start='Jun 1, 2025', duration=14, color='tab:blue'),
    gp.Task('Task H', start='Jun 1, 2025', finish='Jun 25, 2025', color='tab:blue')
]

# Make plot
gp.gantt(tasks, show_durations=True, fname='example.png')
```
