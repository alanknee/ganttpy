```python
import ganttpy as gp
```

```python
# Define the tasks we want to plot
tasks = [
    gp.Task('Task A', start='Feb 3, 2025', end='Feb 20, 2025', color='tab:red'),
    gp.Task('Task B', start='Feb 15 2025', end='Mar 2, 2025', color='tab:red'),
    gp.Task('Task C', start='Feb 8, 2025', end='Mar 6, 2025', color='tab:blue'),
    gp.Task('Task D', start='Feb 25, 2025', end='Mar 2, 2025', color='tab:green'),
    gp.Task('Task E', start='Mar 6, 2025', end='Mar 25, 2025', color='tab:orange'),
    gp.Task('Task F', start='Mar 20, 2025', duration=10, color='tab:blue'),
]

# Make plot
gp.gantt(tasks, order='listed', date_freq='5D', fname='example.png')
```
